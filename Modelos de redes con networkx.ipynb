## *Nodos*

Para agregar un nodo a la vez se emplea la siguiente función:

```python
nx.draw_networkx(G)
G.add_node(1)
```

De igual manera agregar nodos desde cualquier contenedor iterable , como una lista

```python
nx.draw_networkx(G)
G.add_nodes_from([2 , 3])
```

Podemos incorporar los nodos de un gráfico con otro, de la siguiente manera

```python
nx.draw_networkx(G)
G.add_nodes_from(H)
```

G ahora contiene los nodos de H como nodos de G. Por el contrario, podemos usar el gráfico H como un nodo en G.

```python
nx.draw_networkx(G)
G.add_node(H)
```

**Atributos de nodos**

```python
nx.draw_networkx(G)
G.add_node(1, time='5pm')
G.add_nodes_from([3], time='2pm')
G.nodes[1]
```

```python
nx.draw_networkx(G)
G.nodes[1]['room'] = 714
G.nodes.data()
```
